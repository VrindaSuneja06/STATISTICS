Question 1
What is hypothesis testing in statistics?
Hypothesis testing is a statistical method used to make decisions or draw conclusions about a population based on sample data. It involves formulating a null hypothesis and an alternative hypothesis, selecting a significance level, calculating a test statistic, and deciding whether to reject or fail to reject the null hypothesis.

Question 2
What is the null hypothesis, and how does it differ from the alternative hypothesis?
The null hypothesis (H₀) states that there is no effect, no difference, or no relationship in the population.
The alternative hypothesis (H₁) states that there is an effect, difference, or relationship.
The null hypothesis is tested for possible rejection, while the alternative hypothesis is accepted if sufficient evidence exists against the null hypothesis.

Question 3
Explain the significance level in hypothesis testing.
The significance level (α) is the probability of rejecting the null hypothesis when it is actually true. Common values are 0.05 or 0.01.
If the p-value is less than the significance level, the null hypothesis is rejected.

Question 4
What are Type I and Type II errors? Give examples.
Type I Error: Rejecting a true null hypothesis.
Example: Concluding a medicine works when it actually does not.
Type II Error: Failing to reject a false null hypothesis.
Example: Concluding a medicine does not work when it actually does.

Difference between Z-test and T-test.
Z-Test

T test

```python
#7

import numpy as np
from scipy import stats

sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
               50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
               50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
               50.3, 50.4, 50.0, 49.7, 50.5, 49.9]

sample_mean = np.mean(sample_data)
population_mean = 50
std_dev = np.std(sample_data, ddof=1)
n = len(sample_data)

z_stat = (sample_mean - population_mean) / (std_dev / np.sqrt(n))
p_value = 2 * (1 - stats.norm.cdf(abs(z_stat)))

print("Z-statistic:", z_stat)
print("P-value:", p_value)
```

```python
#8
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

data = np.random.normal(loc=50, scale=5, size=1000)

mean = np.mean(data)
std = np.std(data)
confidence_interval = stats.norm.interval(0.95, mean, std/np.sqrt(len(data)))

print("Mean:", mean)
print("95% Confidence Interval:", confidence_interval)

plt.hist(data, bins=30, edgecolor='black')
plt.title("Normal Distribution")
plt.show()
```

```python
#9
import numpy as np
import matplotlib.pyplot as plt

def calculate_z_scores(data):
    mean = np.mean(data)
    std = np.std(data)
    return (data - mean) / std

data = np.random.normal(100, 10, 500)
z_scores = calculate_z_scores(data)

plt.hist(z_scores, bins=30, edgecolor='black')
plt.title("Z-score Distribution")
plt.xlabel("Z-score")
plt.ylabel("Frequency")
plt.show()
```
