1. What is a random variable in probability theory?

```python
#40
import numpy as np
from scipy.stats import norm

# Generate data from a normal distribution
data = np.random.normal(loc=50, scale=5, size=100)

# Sample statistics
sample_mean = np.mean(data)
sample_std = np.std(data, ddof=1)
n = len(data)

# 95% Confidence Interval
confidence_level = 0.95
z = norm.ppf(1 - (1 - confidence_level) / 2)

margin_of_error = z * (sample_std / np.sqrt(n))
lower_bound = sample_mean - margin_of_error
upper_bound = sample_mean + margin_of_error

print("Sample Mean:", sample_mean)
print("95% Confidence Interval:", (lower_bound, upper_bound))
```

```python
#44
import numpy as np
import matplotlib.pyplot as plt

# Generate datasets
a = np.random.randn(100)
b = np.random.randn(100) * 2

# Calculate standard deviations
std_a = np.std(a)
std_b = np.std(b)

print("Standard Deviation of A:", std_a)
print("Standard Deviation of B:", std_b)

# Visualization
plt.hist(a, alpha=0.6, label="Dataset A")
plt.hist(b, alpha=0.6, label="Dataset B")
plt.legend()
plt.title("Comparison of Standard Deviations")
plt.show()
```

```python
#45
import numpy as np
from scipy.stats import iqr

# Generate normal data
data = np.random.normal(0, 1, 100)

# Calculate range and IQR
data_range = np.max(data) - np.min(data)
data_iqr = iqr(data)

print("Range:", data_range)
print("Interquartile Range (IQR):", data_iqr)
```

```python
#46
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import zscore

# Generate data
data = np.random.normal(50, 10, 100)

# Z-score normalization
normalized = zscore(data)

# Plot
plt.plot(normalized, marker='o')
plt.title("Z-score Normalized Data")
plt.xlabel("Index")
plt.ylabel("Z-score")
plt.show()
```

```python
#47
import numpy as np
from scipy.stats import skew, kurtosis

# Generate data
data = np.random.normal(0, 1, 1000)

# Calculate skewness and kurtosis
data_skewness = skew(data)
data_kurtosis = kurtosis(data)

print("Skewness:", data_skewness)
print("Kurtosis:", data_kurtosis)
```
